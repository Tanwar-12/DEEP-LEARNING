# **PERCEPTRON ALGORITHM**

_**The fundamental building block of artificial neural networks is the single-layer perceptron.**_ 
_**This neural network architecture comprises just one input layer and one output layer, making it the simplest form of a neural network.**_

_**Imagine a scenario where you have a dataset consisting of two classes of points, and you want to classify them into two distinct regions.**_
_**The single-layer perceptron can serve as a powerful tool for this task.**_


### **Here's how it works:**

* Input Layer: The input layer of the single-layer perceptron represents the features of the data points. In this scenario, each data point is represented by its coordinates (e.g., x and y).

* Output Layer: The output layer consists of a single neuron that represents the classification decision. It produces an output value based on a weighted sum of the input features.

* Activation Function: The output neuron typically employs an activation function, such as the step function. This function maps the weighted sum of inputs to a binary output value, indicating which region of the data space the input point belongs to.

* Decision Boundary: By adjusting the weights and biases, the single-layer perceptron learns to create a decision boundary that separates the two classes of points in the data space. This decision boundary divides the data into two distinct regions, allowing the network to classify new data points accurately.

### **SINGLE LAYER PERCEPTRON**

* A single-layer perceptron (SLP) is the simplest form of a neural network. It consists of only one layer of artificial neurons, also known as perceptrons. Each perceptron takes multiple inputs, applies weights to them, sums them up, and then passes the result through an activation function to produce an output.
* A single-layer perceptron can only classify linearly separable data. It means it can only classify data points that can be separated by a straight line in a feature space. Because of this limitation, single-layer perceptrons are not suitable for more complex classification tasks like XOR operations.

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

_**LOAD THE DATASET**_

```python
df = pd.read_csv('placement.csv')
```

```python
print(df.shape)
df.head()
```

```python
sns.scatterplot(x=df['cgpa'],y=df['resume_score'],hue=df['placed'])
```

```python
X = df.iloc[:,0:2]
y = df.iloc[:,-1]
```

```python
from sklearn.linear_model import Perceptron
p = Perceptron()
```

```python
p.fit(X,y)
```

```python
p.coef_
```

```python
p.intercept_
```

```python
from mlxtend.plotting import plot_decision_regions
```

```python
plot_decision_regions(X.values, y.values, clf=p, legend=2)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from sklearn.datasets import make_classification
from sklearn.linear_model import Perceptron

```

_**GENERATE SYNTHETC 3D DATA**_ 


```python
X, y = make_classification(n_samples=100, n_features=3, n_classes=2, n_clusters_per_class=1, n_redundant=0, random_state=42)
```

```python
# Initialize and train the perceptron
perceptron = Perceptron(max_iter=100, eta0=0.1, random_state = 42)
```

```python
perceptron.fit(X, y)
```

_**PLOT THE 3D SCATTER PLOTP**_


```python
fig = plt.figure(figsize=(8, 6))
ax = fig.add_subplot(111, projection='3d')
ax.scatter(X[:, 0], X[:, 1], X[:, 2], c=y, cmap='viridis', marker='o', edgecolors='k')
ax.set_xlabel('Feature 1')
ax.set_ylabel('Feature 2')
ax.set_zlabel('Feature 3')

# Create a meshgrid to plot decision regions
xx, yy = np.meshgrid(np.linspace(X[:, 0].min(), X[:, 0].max(), 100),
                     np.linspace(X[:, 1].min(), X[:, 1].max(), 100))
zz = -(perceptron.coef_[0][0] * xx + perceptron.coef_[0][1] * yy + perceptron.intercept_) / perceptron.coef_[0][2]

# Plot decision regions
ax.plot_surface(xx, yy, zz, alpha=0.3, cmap='viridis')
plt.title('Perceptron Decision Boundary in 3D')
plt.show()
```
